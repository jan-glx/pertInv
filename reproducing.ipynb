```python
import numpy as np
import pandas as pd
import gzip
import scipy.io
```

```python
# load cell names
cell_names = pd.read_csv(gzip.open('data/GSM2396859_k562_tfs_13_cellnames.csv.gz','rb'), names=['cell_id','cell'], header=0)
tmp = cell_names['cell'].str.extract('(.*)_(.*_.*)', expand=True)
tmp.columns = ['cell barcode', 'batch']
cell_names = pd.concat([cell_names,tmp], axis=1)
cell_names.head()
```

```python
# load gene names
gene_names = pd.read_csv(gzip.open('data/GSM2396859_k562_tfs_13_genenames.csv.gz','rb'), names=['gene_id','gene name'], header=0)
gene_names.head()
```

```python
cell_guide_map = pd.read_csv(gzip.open('data/GSM2396859_k562_tfs_13_cbc_gbc_dict.csv.gz','rb'), 
                             header=None, names=['cell'], index_col=[0])
cell_guide_map.index.rename('guide', inplace=True)
cell_guide_map = pd.DataFrame(cell_guide_map['cell'].str.split(', ', expand=True).stack().reset_index(level=1, drop=True))
cell_guide_map.columns = [ 'cell']
cell_guide_map.reset_index(inplace=True)
cell_guide_map.head()
```

```python
cell_names = pd.merge(cell_guide_map, cell_names, on='cell', how='right')
```

```python
cell_names.groupby("cell barcode").count().groupby("guide").count()
```

```python
cell_names.groupby("cell").count().groupby("guide").count()
```
